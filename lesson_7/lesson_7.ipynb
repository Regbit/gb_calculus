## Task #6

### $$U=x^2y+\frac{1}{3}y^3+2x^2+3y^2-1$$

### $$\begin{cases}
   U'_{x}=2xy+4x=2x(y+2)=0, \\
   U'_{y}=x^2 + y^2 + 6y = 0
 \end{cases}$$

### $$\begin{cases}
   2x(y+2)=0, \\
   x=\pm\sqrt{-y^2 - 6y}
 \end{cases}$$

### $$\pm2\sqrt{-y^2 - 6y}\cdot(y+2)=0$$
### $$y_1 = 0, x_1 = 0 \rightarrow A(0,0)$$
### $$y_2 = -6, x_2 = 0 \rightarrow B(-6,0)$$
### $$y_3 = -2, x_3 = \sqrt{8} \rightarrow C(-2,\sqrt{8})$$
### $$y_4 = -2, x_4 = -\sqrt{8} \rightarrow D(-2,-\sqrt{8})$$

### $$U''_{xx}=2y+4$$
### $$U''_{xy}=2x$$
### $$U''_{yx}=2x$$
### $$U''_{yy}=2y+6$$

### $$\Delta_{1}=U''_{xx}=2y+4$$

### $$\Delta_{2}=\begin{vmatrix}
U''_{xx} & U''_{xy}\\ 
U''_{yx} & U''_{yy}\\
\end{vmatrix}=U''_{xx}\cdot U''_{yy}-(U''_{xy})^2=(2y+4)(2y+6)-4x^2=4y^2+12y+8y+24-4x^2=4(y^2+5y+6-x^2)$$


Для точки $A(0,0)$: $\Delta_{1}=4>0$; $\Delta_{2}=24>0 \rightarrow A$ - точка минимума

Для точки $B(-6,0)$: $\Delta_{1}=4>0$; $\Delta_{2}=-144<0 \rightarrow B$ - "седловая" точка

Для точки $C(-2,\sqrt{8})$: $\Delta_{1}=2\sqrt{8}+4>0$; $\Delta_{2}=40 + 20\sqrt{8}>0 \rightarrow C$ - точка минимума

Для точки $D(-2,-\sqrt{8})$: $\Delta_{1}=-2\sqrt{8}+4<0$; $\Delta_{2}=40 - 20\sqrt{8}<0 \rightarrow D$ - "седловая" точка

```python
import pylab
from mpl_toolkits.mplot3d import Axes3D
import numpy

def makeData ():
    x = numpy.arange (-10, 10, 0.1)
    y = numpy.arange (-10, 10, 0.1)
    xgrid, ygrid = numpy.meshgrid(x, y)

    Ugrid = xgrid**2*ygrid + (1/3)*ygrid**3 + 2*xgrid**2 + 3*ygrid**2 -1
    return xgrid, ygrid, Ugrid

x, y, U = makeData()

fig = pylab.figure()
axes = Axes3D(fig)

axes.plot_surface(x, y, U)

pylab.show()
```
